```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sys
%config InlineBackend.figure_format = 'retina'
plt.rcParams['font.family'] = 'Times New Roman'
sys.path.insert(0, '/Users/james/Desktop/DRL_Hedging')
```

```python
sac_pnl = np.load("./result_analysis/sac_pnl_sabr_3e-3.npy")
ddpg_pnl = np.load("./result_analysis/ddpg_pnl_sabr_3e-3.npy")
delta_hedging_pnl = np.load("./result_analysis/delta_hedging_pnl_sabr_3e-3.npy")
```

```python
sac_costs = np.load("./result_analysis/sac_costs_sabr_3e-3.npy")
ddpg_costs = np.load("./result_analysis/ddpg_costs_sabr_3e-3.npy")
delta_hedging_costs = np.load("./result_analysis/delta_hedging_costs_sabr_3e-3.npy")
```

```python
sac_rewards = np.load("./result_analysis/sac_rewards_sabr_3e-3.npy")
ddpg_rewards = np.load("./result_analysis/ddpg_rewards_sabr_3e-3.npy")
delta_hedging_rewards = np.load("./result_analysis/delta_hedging_rewards_sabr_3e-3.npy")
```

```python
print("SAC PnL and its std: {:.2f}, {:.2f}".format(sac_pnl.mean(), sac_pnl.std()))
print("DDPG PnL and its std: {:.2f}, {:.2f}".format(ddpg_pnl.mean(), ddpg_pnl.std()))
print("DH PnL and its std: {:.2f}, {:.2f}".format(delta_hedging_pnl.mean(), delta_hedging_pnl.std()))
```

```python
print("SAC Cost and its std: {:.2f}, {:.2f}".format(sac_costs.mean(), sac_costs.std()))
print("DDPG Cost and its std: {:.2f}, {:.2f}".format(ddpg_costs.mean(), ddpg_costs.std()))
print("DH Cost and its std: {:.2f}, {:.2f}".format(delta_hedging_costs.mean(), delta_hedging_costs.std()))
```

```python
print("SAC rewards and its std: {:.2f}, {:.2f}".format(sac_rewards.mean(), sac_rewards.std()))
print("DDPG rewards and its std: {:.2f}, {:.2f}".format(ddpg_rewards.mean(), ddpg_rewards.std()))
print("DH rewards and its std: {:.2f}, {:.2f}".format(delta_hedging_rewards.mean(), delta_hedging_rewards.std()))
```

```python
# KDE plot
sns.kdeplot(sac_rewards, label='SAC', color='black',fill=False)
sns.kdeplot(ddpg_rewards, linestyle='dashed', label='DDPG',color='black', fill=False)
sns.kdeplot(delta_hedging_rewards, linestyle='dotted', label='Delta', color='black',fill=False)

plt.title('KDE total Rewards of DRL Policy and Delta Hedging Policy')
plt.xlabel('Total Rewards')
plt.legend()
plt.show()
```

```python
# KDE plot
sns.kdeplot(sac_costs, label='SAC', color='black',fill=False)
sns.kdeplot(ddpg_costs, linestyle='dashed', label='DDPG',color='black', fill=False)
sns.kdeplot(delta_hedging_costs, linestyle='dotted', label='Delta', color='black',fill=False)

plt.title('KDE total Trading Cost of DRL Policy and Delta Hedging Policy')
plt.xlabel('Total Trading Cost')
plt.legend()
plt.show()
```

```python
# KDE plot
sns.kdeplot(sac_pnl, label='SAC', color='black',fill=False)
sns.kdeplot(ddpg_pnl, linestyle='dashed', label='DDPG',color='black', fill=False)
sns.kdeplot(delta_hedging_pnl, linestyle='dotted', label='Delta', color='black',fill=False)

plt.title('KDE total P&L of DRL Policy and Delta Hedging Policy')
plt.xlabel('Total P&L')
plt.legend()
plt.show()
```
